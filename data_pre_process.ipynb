```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

def split_and_save_data(file_path, output_dir):
    """
    file_path: 분할할 파일의 전체 경로
    output_dir: 분할된 데이터를 저장할 디렉토리 경로
    """
    # 파일 경로와 이름 처리
    parts = file_path.split('/')
    category = parts[-2] # 예: 'normal'
    file_name = parts[-1].replace('.csv', '') # 예: 'tower_final'
    base_name = f"{category}_{file_name}" # 예: 'normal_tower_final'
    
    # 데이터 로드
    data = pd.read_csv(file_path)
    
    # 데이터 분할: train 80%, val 10%, test 10%
    train_val, test = train_test_split(data, test_size=0.1, random_state=42)
    train, val = train_test_split(train_val, test_size=0.1111, random_state=42) # 0.1111 = 10% of 90%
    
    # 파일 저장
    train.to_csv(os.path.join(output_dir, f'{base_name}_train.csv'), index=False)
    val.to_csv(os.path.join(output_dir, f'{base_name}_val.csv'), index=False)
    test.to_csv(os.path.join(output_dir, f'{base_name}_test.csv'), index=False)

def merge_datasets(output_dir, dataset_type):
    """
    output_dir: 분할된 데이터가 저장된 디렉토리 경로
    dataset_type: 합칠 데이터셋 유형 ('train', 'val', 'test')
    """
    merged_data = pd.DataFrame()
    for file in os.listdir(output_dir):
        if file.endswith(f'{dataset_type}.csv'):
            data = pd.read_csv(os.path.join(output_dir, file))
            merged_data = pd.concat([merged_data, data], ignore_index=True)
    
    merged_data.to_csv(os.path.join(output_dir, f'all_{dataset_type}.csv'), index=False)

# 파일 경로 설정 및 실행
base_dir = '/Users/junwonlee/Desktop/LG_2024_2/all_최종2' # 이 경로를 실제 데이터가 있는 디렉토리 경로로 변경하세요.
output_dir = '/Users/junwonlee/Desktop/LG_2024_2/all_input2' # 출력 디렉토리 경로, 적절히 변경하세요.
files = [
    'normal/tower_final.csv',
    'normal/winner_final.csv',
    'random/tower_final.csv',
    'random/winner_final.csv'
]

# 각 파일에 대해 데이터 분할 및 저장 실행
for file_name in files:
    file_path = os.path.join(base_dir, file_name)
    split_and_save_data(file_path, output_dir)

# 분할된 데이터셋 합치기
for dataset_type in ['train', 'val', 'test']:
    merge_datasets(output_dir, dataset_type)
```

```python
# Let's calculate the performance without rounding the values, using a 21.74% error tolerance.

answers = [3.7, 4.25, 3.33, 4.5, 4.6, 3, 4.75, 4.5]
predictions = [3.66, 4.0, 3.0, 3.4, 3.6, 4.0, 3.9, 5.0]

tolerance_percentage = 25

# Calculate performance without rounding
correct_count = 0
for answer, prediction in zip(answers, predictions):
    error_margin = abs(answer - prediction) / answer
    if error_margin <= tolerance_percentage / 100:
        correct_count += 1

performance = (correct_count / len(answers)) * 100
performance
```
